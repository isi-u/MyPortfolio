```python
#The following libraries are required to run this notebook
#Importing the

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
#Load the car.csv dataset using the panda.read_csv function
#used the r prefix to avoid the file path from executing as a string
car = pd.read_csv(r"C:\Users\Israel\Downloads\cars.csv\data.csv")
car.head(10)
```

```python
#to know the data types for each row
car.dtypes
```

```python
#to know the column names
# this converts it from a pandas list to a numpy array #car.columns.values
car.columns
```

```python
#Analytical summary of the dataset, really important for data scientist
car.describe(include='all')
```

```python
#histogram of all the different pieces in our dataframe
car.hist(figsize=(20,30))
```

```python
# Relationship between categorical and continous variable
sns.boxplot(x="Vehicle Size", y="Engine HP", data=car)
```

```python
#engine horsepower gets bigger as time goes on
sns.pairplot(car)
```

```python
#Dropping irrelevant columns 
car = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
car.head(7)
```

```python
#renaming column names
car = car.rename(columns = {"Engine HP":"HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels":"Driven Mode", "Highway MPG": "MPG-H"})
car.head(7) 
#renaming the column names made readability easier
```

```python
#total number of rows and columns
car.shape
#shows relationship between pandas and numpy
```

```python
#rows with duplicate
DuplicatedRows = car[car.duplicated()]
print("no. of duplicate rows", DuplicatedRows.shape) #the .shape is used to give a summarization of the data
```

```python
#to count the rows so as to compare after dropping duplicates
car.count()
```

```python
#drop duplicates
car = car.drop_duplicates()
car.count()
```

```python
#find the null values
print(car.isnull().sum()) #the sum function is used to sum the null values to give the total count of null values.
```

```python
#drop the missing values
car=car.dropna()
car.count ()
```

```python
#this should give zero null values
print(car.isnull().sum())
```

```python
#finding outliers using the boxplot using msrp which is the price
sns.boxplot(x=car["MSRP"])
#the high end price are the expensive cars
```

```python
#finding outliers using the boxplot using horse power
sns.boxplot(x=car["HP"])
#the farther it is from the the actual plot the greater the horse power
```

```python
#plotting histogram for the number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5), color='red')
#for readability
plt.title('Number_of_cars_made_by_each_brand')
plt.ylabel('Number_of_cars')
plt.xlabel('Brand')
```

```python
#relationship between variable
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(data=c,cmap='BrBG', annot=True)
c
#the closer it is to one, if positive it is positively correlated and vice versa
```
